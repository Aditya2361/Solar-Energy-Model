```python
from solar_model import load_solar_data, calculate_stats, plot_daily_output, compare_locations, plot_comparison


# SETTINGS - edit these for your own scenario

panel_area = 1.7            # m² per panel
efficiency = 0.20           # panel efficiency (decimal)
performance_ratio = 0.78    # real-world system losses
electricity_price = 0.315   # $ per kWh
number_of_panels = 16       # panels in a full system
system_cost = 5500          # installed system cost ($)


# LOAD & CALCULATE - Sydney

df_sydney = load_solar_data('SydneyData.csv', panel_area, efficiency, performance_ratio)
stats_sydney = calculate_stats(df_sydney, electricity_price, number_of_panels, system_cost)
plot_daily_output(df_sydney, 'Sydney')


# LOAD & CALCULATE - Alice Springs

df_alice = load_solar_data('AliceSpringsData.csv', panel_area, efficiency, performance_ratio)
stats_alice = calculate_stats(df_alice, electricity_price, number_of_panels, system_cost)
plot_daily_output(df_alice, 'Alice Springs')


# COMPARE

comparison_table = compare_locations('Sydney', stats_sydney, 'Alice Springs', stats_alice)
print(comparison_table)

plot_comparison(df_sydney, 'Sydney', df_alice, 'Alice Springs')
```
